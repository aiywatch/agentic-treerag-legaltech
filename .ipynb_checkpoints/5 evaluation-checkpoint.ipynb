```python
import os
import json
import pandas as pd
from tqdm import tqdm
from openai import OpenAI
import re
```

```python
client = OpenAI(api_key=GITHUB_KEY, base_url=GITHUB_URL)
```

```python

def calculate_clause_metrics(expected_str, predicted_str):
    # ป้องกันกรณีค่าว่าง (NaN)
    if pd.isna(expected_str): expected_str = ""
    if pd.isna(predicted_str): predicted_str = ""
    
    # แยกข้อกฎหมายด้วยลูกน้ำ (,) ตัดช่องว่าง 
    expected_list = [x.strip() for x in str(expected_str).split(',') if x.strip()]
    predicted_list = [x.strip() for x in str(predicted_str).split(',') if x.strip()]
    
    # กรณีไม่มีข้อกฎหมายทั้งคู่ (คำถามที่ไม่ต้องอิงมาตรา)
    if not expected_list and not predicted_list:
        return 1.0, 1.0, 1.0 
    
    # กรณีเฉลยมีแต่ทายไม่มี หรือทายมาแต่เฉลยไม่มี
    if not expected_list or not predicted_list:
        return 0.0, 0.0, 0.0
        
    true_positives = 0
    matched_preds = set()
    matched_exps = set()
    
    # ฟังก์ชันช่วยสกัดเอาเฉพาะ "ข้อหลัก" (ตัดข้อย่อยทิ้งทั้งหมด)
    def get_base_clause(clause_str):
        # จับเฉพาะคำว่า "ข้อ" เว้นวรรคตามด้วย "ตัวเลข" เช่น ข้อ 53
        match = re.match(r'^(ข้อ\s*\d+)', clause_str)
        return match.group(1) if match else clause_str
        
    # ลูปเช็คความสอดคล้องแบบ Smart Match (เทียบแค่ข้อหลัก)
    for pred in predicted_list:
        base_pred = get_base_clause(pred)
        for exp in expected_list:
            base_exp = get_base_clause(exp)
            
            # เงื่อนไขที่ให้ผ่าน: เหมือนกันเป๊ะที่ข้อหลัก (ไม่สนว่าข้อย่อยคืออะไร)
            if base_pred == base_exp:
                
                # ถ้าเจอกันแล้ว และยังไม่เคยถูกนับคะแนน ให้บวกคะแนนได้เลย
                if pred not in matched_preds and exp not in matched_exps:
                    true_positives += 1
                    matched_preds.add(pred)
                    matched_exps.add(exp)
                    break # เจอคู่แล้ว ข้ามไปเช็คข้อถัดไป
    
    # คำนวณ Precision, Recall, F1
    precision = true_positives / len(predicted_list) if len(predicted_list) > 0 else 0.0
    recall = true_positives / len(expected_list) if len(expected_list) > 0 else 0.0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0.0
    
    return precision, recall, f1
    
# ==========================================
# 🛠️ ฟังก์ชันย่อย 2: ให้คะแนน "คำตอบสรุป" (LLM-as-a-Judge)
# ==========================================
def evaluate_answer_with_llm(query, context, generated_answer, expected_answer):
    eval_prompt = f"""
    คุณคือกรรมการประเมินผลระบบ AI สกัดข้อกฎหมาย (Legal RAG Evaluator)
    จงประเมินผลลัพธ์ต่อไปนี้อย่างเป็นกลาง และตอบกลับเป็นรูปแบบ JSON เท่านั้น
    
    [ข้อมูล]
    คำถาม: {query}
    ข้อมูลอ้างอิงที่หามาได้ (Context): {context}
    คำตอบที่ AI สร้างขึ้น (Generated): {generated_answer}
    คำตอบที่ถูกต้อง (Ground Truth): {expected_answer}
    
    [เกณฑ์การประเมิน]
    1. faithfulness (0 หรือ 1): คำตอบที่สร้างขึ้นมาจาก Context จริงๆ หรือไม่? (ถ้ามโนให้ 0, ถ้าอิงจาก Context ให้ 1)
    2. relevance_score (1 ถึง 5): คำตอบตรงประเด็นและมีความหมายสอดคล้องกับ Ground Truth แค่ไหน? (5 = สมบูรณ์แบบ, 1 = ตอบผิด/ไม่ตรงคำถามเลย)
    
    [รูปแบบ Output JSON]
    {{
        "faithfulness": 1,
        "relevance_score": 4,
        "reason": "คำอธิบายสั้นๆ ว่าทำไมถึงหักคะแนนหรือให้คะแนนเต็ม"
    }}
    """
    try:
        response = client.chat.completions.create(
            model=llm_model,
            messages=[{"role": "user", "content": eval_prompt}],
            response_format={"type": "json_object"},
            temperature=0.0
        )
        return json.loads(response.choices[0].message.content)
    except Exception as e:
        return {"faithfulness": 0, "relevance_score": 0, "reason": f"API Error: {str(e)}"}

```

```python
# ==========================================
# 🚀 ฟังก์ชันหลัก: รับ JSON (ข้อสอบ) และ DF (คำตอบ)
# ==========================================
def run_evaluation_from_two_sources(json_filepath, df_answers):
    # 1. โหลดคำถามและเฉลยจากไฟล์ JSON
    print(f"📂 กำลังโหลดชุดข้อสอบจาก: {json_filepath}")
    with open(json_filepath, 'r', encoding='utf-8') as f:
        questions_data = json.load(f)
    
    # แปลง JSON เป็น DataFrame เพื่อให้จัดการง่ายขึ้น
    df_questions = pd.DataFrame(questions_data)
    
    # แปลง List ของ reference_clause ให้เป็น String คั่นด้วยลูกน้ำ (เตรียมไว้เทียบ)
    df_questions['expected_clause'] = df_questions['reference_clause'].apply(lambda x: ", ".join(x) if isinstance(x, list) else str(x))
    
    print(f"📂 กำลังนำข้อสอบ {len(df_questions)} ข้อ มาจับคู่กับคำตอบจาก DataFrame...")
    
    # 2. นำ DataFrame ข้อสอบ มา Join กับ DataFrame คำตอบ (ดึงมาเฉพาะคอลัมน์ที่จำเป็น)
    # ใช้ 'id' เป็นกุญแจในการเชื่อมข้อมูล
    df_merged = pd.merge(
        df_questions, 
        df_answers[['id', 'naive_rag_clause', 'naive_rag_answer', 'retrieved_context']], 
        on='id', 
        how='left'
    )
    
    eval_results = []
    
    # 3. รันประเมินผลทีละข้อ
    for index, row in tqdm(df_merged.iterrows(), total=df_merged.shape[0], desc="Evaluating"):
        
        # ประเมินข้อกฎหมาย
        p, r, f1 = calculate_clause_metrics(row['expected_clause'], row['naive_rag_clause'])
        
        # ประเมินคำตอบสรุปด้วย LLM
        llm_eval = evaluate_answer_with_llm(
            query=row['query'],
            context=row['retrieved_context'],
            generated_answer=row['naive_rag_answer'],
            expected_answer=row['expected_answer']
        )
        
        eval_results.append({
            "id": row['id'],
            "clause_precision": p,
            "clause_recall": r,
            "clause_f1": f1,
            "ans_faithfulness": llm_eval.get('faithfulness', 0),
            "ans_relevance": llm_eval.get('relevance_score', 0),
            "eval_reason": llm_eval.get('reason', '')
        })
        
    return pd.DataFrame(eval_results), df_merged

# ==========================================
# 📂 ส่วนการเรียกใช้งานฟังก์ชัน
# ==========================================
```

```python


# ==========================================
# 📊 ฟังก์ชันหลัก: วิเคราะห์สถิติและให้ LLM เขียนสรุป
# ==========================================
def generate_academic_summary(csv_filepath, model_name='Model'):
    print(f"📂 กำลังอ่านข้อมูลผลการประเมินจาก: {csv_filepath}")
    
    try:
        df = pd.read_csv(csv_filepath)
    except FileNotFoundError:
        return "❌ ไม่พบไฟล์ CSV กรุณาตรวจสอบชื่อไฟล์อีกครั้ง"

    # 1. คำนวณสถิติภาพรวม (Overall Metrics)
    overall_stats = {
        "Total Questions": len(df),
        "Avg Clause Precision": df['clause_precision'].mean(),
        "Avg Clause Recall": df['clause_recall'].mean(),
        "Avg Clause F1": df['clause_f1'].mean(),
        "Avg Faithfulness (%)": df['ans_faithfulness'].mean() * 100,
        "Avg Relevance (out of 5)": df['ans_relevance'].mean()
    }

    # 2. คำนวณสถิติแยกตามระดับความยาก (ถ้ามีคอลัมน์ difficulty)
    difficulty_stats_str = ""
    if 'difficulty' in df.columns:
        diff_group = df.groupby('difficulty')[['clause_f1', 'ans_relevance']].mean()
        difficulty_stats_str = "\nสถิติแยกตามความยาก (เฉลี่ย):\n"
        for index, row in diff_group.iterrows():
            difficulty_stats_str += f"- ระดับ {index}: Clause F1 = {row['clause_f1']:.2f}, Relevance = {row['ans_relevance']:.2f}\n"

    # 3. เตรียมข้อมูลสถิติเพื่อส่งให้ LLM
    stats_text = f"""
    [ข้อมูลสถิติภาพรวมของ {model_name}]
    - จำนวนข้อสอบทั้งหมด: {overall_stats['Total Questions']} ข้อ
    - ความแม่นยำในการดึงข้อกฎหมาย (Clause Precision): {overall_stats['Avg Clause Precision']:.2f}
    - ความครอบคลุมในการดึงข้อกฎหมาย (Clause Recall): {overall_stats['Avg Clause Recall']:.2f}
    - ค่า F1-Score ของข้อกฎหมาย (Clause F1): {overall_stats['Avg Clause F1']:.2f}
    - ความซื่อตรงของคำตอบ (Faithfulness): {overall_stats['Avg Faithfulness (%)']:.2f}%
    - ความตรงประเด็นของคำตอบ (Relevance): {overall_stats['Avg Relevance (out of 5)']:.2f} จาก 5.00
    {difficulty_stats_str}
    """

    print("🧠 กำลังให้ LLM วิเคราะห์และเขียนสรุปผลระดับวิชาการ...")

    # 4. สร้าง Prompt ให้ LLM เขียนสรุป
    prompt = f"""
    คุณคือนักวิจัยด้าน AI และ Legal Technology ที่กำลังเขียนบทสรุปผลการทดลอง (Discussion & Conclusion) สำหรับวิทยานิพนธ์
    
    จงอ่านข้อมูลสถิติการประเมินผลระบบ "{model_name}" (ระบบค้นหาข้อกฎหมายแบบพื้นฐาน) ด้านล่างนี้:
    
    {stats_text}
    
    หน้าที่ของคุณ:
    เขียนสรุปวิเคราะห์ผลการทำงานของ {model_name} โดยใช้ภาษาทางการเชิงวิชาการ (Academic Tone) 
    ความยาวประมาณ 2-3 ย่อหน้า โดยให้ครอบคลุมประเด็นต่อไปนี้:
    1. ภาพรวมประสิทธิภาพของระบบ (จากค่า F1, Faithfulness และ Relevance)
    2. จุดเด่นหรือจุดอ่อนที่สังเกตเห็น (เช่น ทำได้ดีในข้อระดับไหน และมีข้อจำกัดในระดับไหน)
    3. ข้อเสนอแนะสั้นๆ ว่าทำไมโครงสร้างแบบ Naive RAG ถึงอาจจะไม่เพียงพอสำหรับเอกสารกฎหมายที่มีความซับซ้อน (เพื่อปูทางไปสู่การเสนอ TreeRAG ต่อไป)
    """

    try:
        # เรียกใช้ LLM
        response = client.chat.completions.create(
            model=llm_model,
            messages=[{"role": "user", "content": prompt}],
            temperature=0.3 # ใช้ temperature ต่ำเพื่อให้ได้ภาษาทางการและมีเหตุผล
        )
        
        summary_text = response.choices[0].message.content.strip()
        
        # 5. พิมพ์ผลลัพธ์และเซฟลงไฟล์ Text
        print("\n" + "="*50)
        print("📝 สรุปผลการประเมิน (AI Generated Academic Summary)")
        print("="*50)
        print(summary_text)
        print("="*50)
        
        with open("evaluation_summary_report.txt", "w", encoding="utf-8") as text_file:
            text_file.write(stats_text + "\n\n" + "="*30 + "\n\n" + summary_text)
            
        print("\n✅ บันทึกคำอธิบายสรุปผลสำเร็จที่ไฟล์: evaluation_summary_report.txt")
        return summary_text

    except Exception as e:
        print(f"❌ เกิดข้อผิดพลาดในการเรียก API: {e}")
        return None
```

```python
try:
    # โหลดไฟล์คำตอบที่ได้จาก Naive RAG (มองว่าเป็น DF ตามโจทย์)
    df_generated_answers = pd.read_csv("results/naive_rag_results.csv")
    
    # เรียกใช้ฟังก์ชันโดยส่ง JSON และ DF เข้าไป
    df_scores, df_full_data = run_evaluation_from_two_sources(QUESTION_JSON_PATH, df_generated_answers)
    
    # นำคะแนนมารวมกับข้อมูลตั้งต้น
    df_final_report = pd.merge(df_full_data, df_scores, on="id")
    
    # บันทึกไฟล์สรุป
    df_final_report.to_csv("evaluation_report_naive_rag.csv", index=False, encoding='utf-8-sig')
    print("\n✅ ประเมินผลเสร็จสิ้น! บันทึกรายงานที่: evaluation_report_naive_rag.csv")

except Exception as e:
    print(f"❌ เกิดข้อผิดพลาด: {e}")
```

```python
import pandas as pd

# ==========================================
# 🎯 ขั้นตอนที่ 1: อ่านไฟล์คำตอบและเตรียมข้อมูลให้ตรงฟอร์แมต
# ==========================================
try:
    print("📂 1. กำลังโหลดไฟล์คำตอบของ TreeRAG (legaltreerag_ans.csv)...")
    # อ่านไฟล์ผลลัพธ์ TreeRAG ที่คุณอัปโหลดมา
    # df_treerag = pd.read_csv("legaltreerag_ans_context.csv")
    df_treerag = pd.read_excel("legaltreerag_ans_context.xlsx")
    
    # 🛠️ แปลงชื่อคอลัมน์จาก TreeRAG ให้ตรงกับที่ฟังก์ชันหลักต้องการ
    df_treerag = df_treerag.rename(columns={
        "Q_ID": "id",
        "Generated Clauses": "naive_rag_clause", # แม้ชื่อตัวแปรจะเป็น naive แต่มันคือคำตอบจาก treeRAG
        "Generated Answer": "naive_rag_answer",
        "Retrieved Context": "retrieved_context" # 🟢 แมปคอลัมน์ Context ที่เราเพิ่งเพิ่มเข้าไป
    })
    
    # 🛠️ เติมคอลัมน์ retrieved_context เผื่อกรณีรันไฟล์เก่าที่ยังไม่มีคอลัมน์นี้ จะได้ไม่ Error
    if 'retrieved_context' not in df_treerag.columns:
        print("⚠️ คำเตือน: ไม่พบคอลัมน์ 'Retrieved Context' ในไฟล์ CSV (อาจทำให้คะแนน Faithfulness ตก)")
        df_treerag['retrieved_context'] = "ระบบ TreeRAG ไม่ได้บันทึกข้อมูล Context สำหรับข้อนี้"

    print("\n🚀 2. เริ่มกระบวนการประเมินผล (เปรียบเทียบคำตอบกับ eval_questions2.json)...")
    # เรียกใช้ฟังก์ชันประเมินผลที่เราสร้างไว้ (ใช้ dataframe ที่แก้ชื่อคอลัมน์แล้ว)
    df_scores, df_full_data = run_evaluation_from_two_sources(QUESTION_JSON_PATH, df_treerag)
    
    # รวมตารางคะแนนเข้ากับข้อมูลตั้งต้น
    df_final_report = pd.merge(df_full_data, df_scores, on="id")
    
    # บันทึกเป็นไฟล์ CSV สำหรับผลของ TreeRAG
    output_csv = "evaluation_report_treerag.csv"
    df_final_report.to_csv(output_csv, index=False, encoding='utf-8-sig')
    print(f"\n✅ บันทึกรายงานการประเมินผล TreeRAG สำเร็จที่ไฟล์: {output_csv}")

# ==========================================
# 📝 ขั้นตอนที่ 2: สร้างคำอธิบายสรุปผลลัพธ์
# ==========================================
    print("\n" + "="*50)
    print("📊 3. กำลังสร้างสรุปผลลัพธ์ทางวิชาการ (Academic Summary)...")
    print("="*50)
    
    # เรียกใช้ฟังก์ชันให้ LLM สรุปผลจากไฟล์ CSV ที่เพิ่งบันทึกไป
    treerag_summary = generate_academic_summary(output_csv, 'TreeRag')

except FileNotFoundError as e:
    print(f"\n❌ เกิดข้อผิดพลาด: ไม่พบไฟล์ {e.filename} กรุณาตรวจสอบว่าไฟล์อยู่ในโฟลเดอร์เดียวกับ Notebook แล้ว")
except Exception as e:
    print(f"\n❌ เกิดข้อผิดพลาดที่ไม่คาดคิด: {e}")
```
